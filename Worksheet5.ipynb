```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(X, Y, W, learning_rate, num_iterations):
    m = X.shape[0]
    for i in range(num_iterations):
        predictions = np.dot(X, W)
        error = predictions - Y
        gradient = np.dot(X.T, error) / m
        W = W - learning_rate * gradient
    return W

def simple_train_test_split(X, Y, test_size=0.2):
    split_index = int(X.shape[0] * (1 - test_size))
    X_train, X_test = X[:split_index], X[split_index:]
    Y_train, Y_test = Y[:split_index], Y[split_index:]
    return X_train, X_test, Y_train, Y_test

def model_pipeline(data, test_size=0.2, learning_rate=0.01, iterations=1000):
    X, Y = data[:, :-1], data[:, -1].reshape(-1, 1)
    X_train, X_test, Y_train, Y_test = simple_train_test_split(X, Y, test_size=test_size)
    X_train = (X_train - np.mean(X_train, axis=0)) / np.std(X_train, axis=0)
    X_test = (X_test - np.mean(X_test, axis=0)) / np.std(X_test, axis=0)
    d = X_train.shape[1]
    W = np.random.rand(d, 1)
    W = gradient_descent(X_train, Y_train, W, learning_rate, iterations)
    Y_pred = np.dot(X_test, W)
    rmse = np.sqrt(np.mean((Y_test - Y_pred) ** 2))
    ss_total = np.sum((Y_test - np.mean(Y_test)) ** 2)
    ss_residual = np.sum((Y_test - Y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)
    print("Root Mean Squared Error (RMSE):", rmse)
    print("R^2 Score:", r2)
    return W

data = np.random.rand(1000, 3)
W_learned = model_pipeline(data, test_size=0.2, learning_rate=0.01, iterations=1000)
print("Learned Weights:\n", W_learned)
```
